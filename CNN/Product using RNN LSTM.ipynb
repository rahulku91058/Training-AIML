```python
import numpy as np
from random import seed
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
from math import sqrt

# Mock random_product_pairs function
def random_product_pairs(n_examples, n_numbers, largest):
    X = np.random.randint(1, largest, (n_examples, n_numbers))
    y = np.prod(X, axis=1)  # Product of the numbers as the target
    return X, y

# Mock invert function (if required, replace with correct logic)
def invert(value, n_numbers, largest):
    return int(value)

# Generate training data
seed(1)
n_examples1 = 100
n_examples2 = 10
n_numbers = 2
largest = 100

# Define LSTM configuration
n_batch = 1
n_epoch = 200

# Create LSTM model
model = Sequential()
model.add(LSTM(50, input_shape=(n_numbers, 1)))
model.add(Dense(5))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer="adam")

# Train LSTM
for _ in range(n_epoch):
    X, y = random_product_pairs(n_examples1, n_numbers, largest)
    XX = X.reshape(n_examples1, n_numbers, 1)  # Reshape for LSTM
    model.fit(XX, y, epochs=1, batch_size=n_batch, verbose=2)

# Evaluate on some new patterns
X, y = random_product_pairs(n_examples2, n_numbers, largest)
myX = X  # Original input
myy = y  # Target values
XX = X.reshape(n_examples2, n_numbers, 1)  # Reshape for LSTM

result = model.predict(XX, batch_size=n_batch, verbose=0)

# Calculate RMSE
expected = [invert(x, n_numbers, largest) for x in y]
predicted = [invert(x, n_numbers, largest) for x in result[:, 0]]
rmse = sqrt(mean_squared_error(expected, predicted))
print("RMSE: %f" % rmse)

# Show some examples
for i in range(min(10, n_examples2)):
    error = expected[i] - predicted[i]
    print("%d X %d = %d, Predicted=%d (err=%d)" % (myX[i][0], myX[i][1], expected[i], predicted[i], error))
```
