```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
# Download necessary NLTK data
nltk.download('punkt_tab') # For tokenization
nltk.download('stopwords') # For stopwords
```

```python
# Example text
text="Natural Language Processing is an amazing feild!"
```

```python
# Tokenization
tokens=word_tokenize(text)
```

```python
# Stoppwords removal
stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens if word.lower() not in stop_words]
```

```python
# Output
print("Original Tokens:",tokens)
print("Filtered Tokens:",filtered_tokens)
```

```python
from transformers import AutoModelForCausalLM,AutoTokenizer
```

```python
# Load pre-trained model and tokenizer
model_name="microsoft/DialoGPT-small"
tokrnizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Chatbot interaction
user_input="define AI."
input_ids=tokrnizer.encode(user_input,return_tensors="pt")
```

```python
# Generate response
response_ids=model.generate(input_ids,max_length=15)
response=tokrnizer.decode(response_ids[0],skip_special_tokens=True)
print("ChatBot: ",response)
```
